```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from IPython.display import display

# Dataframe in : https://scikit-learn.org/stable/datasets/toy_dataset.html#diabetes-dataset
url = "https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt"
diabetes_df = pd.read_table(url, sep='\t')
display(diabetes_df.head())
display(diabetes_df.describe())
sns.set(font_scale=2)
sns.set_style('whitegrid')
for feature in diabetes_df.columns[:-1]:
    plt.figure(figsize=(16, 9))
    sns.scatterplot(data=diabetes_df, x=feature, y='Y', hue='Y', palette='cool', legend=False)

X_train, X_test, y_train, y_test = train_test_split(diabetes_df.iloc[:, :-1], diabetes_df['Y'], random_state=11)
linear_regression = LinearRegression()
linear_regression.fit(X=X_train, y=y_train)
predicted = linear_regression.predict(X_test)
expected = y_test
display(metrics.r2_score(expected, predicted))
display(metrics.mean_squared_error(expected, predicted))
diabetes_df.columns

```
